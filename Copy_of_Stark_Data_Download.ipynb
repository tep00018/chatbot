```python
import json
import numpy as np

def convert_to_serializable(obj):
    """Converts NumPy types, tensors, and other objects into serializable formats."""
    if isinstance(obj, np.ndarray):  # Convert NumPy arrays to lists
        return obj.tolist()
    elif isinstance(obj, np.bool_):  # Convert NumPy booleans to Python booleans
        return bool(obj)
    elif isinstance(obj, torch.Tensor):  # Convert PyTorch tensors to lists
        return obj.tolist()
    elif isinstance(obj, dict):  # Recursively process dictionaries
        return {key: convert_to_serializable(value) for key, value in obj.items()}
    elif isinstance(obj, list):  # Recursively process lists
        return [convert_to_serializable(item) for item in obj]
    else:
        return obj  # Default case

if hasattr(skb, "node_info"):
    sample_nodes = list(skb.node_info.keys())[:10]  # First 10 nodes
    sample_data = {node: convert_to_serializable(skb.node_info[node]) for node in sample_nodes}

    print(json.dumps(sample_data, indent=4))  # Pretty-print JSON
else:
    print("No node_info attribute found.")
```

```python
import pandas as pd

# Sample data (your node_df DataFrame)
# Assuming node_df is already loaded into your environment

# Define a function to create searchable text for each row
def create_searchable_text(row):
    text = ""

    # Concatenate columns that are useful for search
    text += f"Title: {row['title']}\n"
    text += f"Review: {row['review']}\n"
    text += f"QA: {row['qa']}\n"
    text += f"ASIN: {row['asin']}\n"
    text += f"Global Category: {row['global_category']}\n"
    text += f"Category: {', '.join(eval(str(row['category']))) if isinstance(row['category'], list) else row['category']}\n"
    text += f"Price: {row['price']}\n"
    text += f"Brand: {row['brand']}\n"
    text += f"Features: {', '.join(eval(str(row['feature']))) if isinstance(row['feature'], list) else row['feature']}\n"
    text += f"Rank: {row['rank']}\n"
    text += f"Description: {', '.join(eval(str(row['description']))) if isinstance(row['description'], list) else row['description']}\n"
    text += f"Details: {row['details']}\n"

    # Optionally, include other columns like 'color', 'brand_name', etc. if relevant
    text += f"Color: {row['color']}\n"
    text += f"Brand Name: {row['brand_name']}\n"
    text += f"Category Name: {row['category_name']}\n"
    text += f"Color Name: {row['color_name']}\n"

    return text.strip()  # Remove any trailing newline

# Apply the function to each row and create a 'searchable_text' column
node_df['searchable_text'] = node_df.apply(create_searchable_text, axis=1)

# Check the result
print(node_df[['node_id', 'searchable_text']].head())
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Load the saved node and edge CSV files
node_df = pd.read_csv('nodes.csv')
edge_df = pd.read_csv('edges.csv')

# Handle missing values
node_df.fillna({'vote': 0, 'details': 'No details', 'rank': '999999 in No Category'}, inplace=True)

# Ensure required text columns exist before applying transformations
text_cols = ['reviewText', 'summary', 'description']
for col in text_cols:
    if col in node_df.columns:
        node_df[col] = node_df[col].astype(str).str.lower().str.replace(r'[^\w\s]', '', regex=True)
    else:
        node_df[col] = ""  # Add missing text columns with empty strings

# Feature extraction using TF-IDF
tfidf = TfidfVectorizer(max_features=100)
text_features = tfidf.fit_transform(node_df['reviewText'] + " " + node_df['summary'] + " " + node_df['description'])

# Convert text features to DataFrame
text_features_df = pd.DataFrame(text_features.toarray(), columns=[f'tfidf_{i}' for i in range(text_features.shape[1])])

# Combine with the main node dataframe
node_df = pd.concat([node_df, text_features_df], axis=1)

# Load dataset split indices correctly
dataset_name = 'amazon'
qa_dataset = load_qa(dataset_name)
idx_split = qa_dataset.get_idx_split()

# Ensure indices align correctly before splitting
node_df.reset_index(drop=True, inplace=True)

# Split data into training, validation, and test sets
train_idx, val_idx = train_test_split(idx_split['train'], test_size=0.2, random_state=42)
test_idx = idx_split['test']

# Convert index lists to actual data selection
train_data = node_df.iloc[train_idx]
val_data = node_df.iloc[val_idx]
test_data = node_df.iloc[test_idx]

# Save preprocessed data
train_data.to_csv('train_data.csv', index=False)
val_data.to_csv('val_data.csv', index=False)
test_data.to_csv('test_data.csv', index=False)
edge_df.to_csv('edges.csv', index=False)

print("Processing and saving complete!")
```

```python
import torch
from torch_geometric.data import Data, DataLoader
from deepseek_v3 import DeepSeekV3

# Load preprocessed data
train_data = pd.read_csv('train_data.csv')
val_data = pd.read_csv('val_data.csv')
test_data = pd.read_csv('test_data.csv')
edges = pd.read_csv('edges.csv')

# Convert data to PyTorch Geometric format
node_features = torch.tensor(train_data.drop(columns=['node_id']).values, dtype=torch.float)
edge_index = torch.tensor(edges.values, dtype=torch.long).t().contiguous()

data = Data(x=node_features, edge_index=edge_index)

# Create DataLoader
train_loader = DataLoader([data], batch_size=32, shuffle=True)

# Initialize DeepSeek-V3 model
model = DeepSeekV3(input_dim=node_features.size(1), hidden_dim=128, output_dim=64)

# Define optimizer and loss function
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = torch.nn.CrossEntropyLoss()

# Training loop
def train():
    model.train()
    for data in train_loader:
        optimizer.zero_grad()
        out = model(data.x, data.edge_index)
        loss = criterion(out, data.y)  # Assuming you have labels in data.y
        loss.backward()
        optimizer.step()

# Evaluation function
def evaluate(loader):
    model.eval()
    correct = 0
    for data in loader:
        out = model(data.x, data.edge_index)
        pred = out.argmax(dim=1)
        correct += int((pred == data.y).sum())
    return correct / len(loader.dataset)

# Train and evaluate
for epoch in range(1, 101):
    train()
    train_acc = evaluate(train_loader)
    val_acc = evaluate(val_loader)
    print(f'Epoch: {epoch:03d}, Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}')

# Test the model
test_acc = evaluate(test_loader)
print(f'Test Accuracy: {test_acc:.4f}')
```
